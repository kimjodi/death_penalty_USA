## 6. Renaming Columns

```python
# Rename columns whose names are too long

df_2.rename(columns = {'Number of White Male Victims' : 'white_male_victims', 'Number of Black Male Victims' : 'black_male_victims', 
                     'Number of Latino Male Victims' : 'latino_male_victims', 'Number of Asian Male Victims' : 'asian_male_victims',
                     'Number of Native American Male Victims' : 'native_amer_male_victims', 'Number of Other Race Male Victims' : 'other_race_male_victims', 'Number of White Female Victims' : 'white_female_victims', 'Number of Black Female Victims' : 'black_female_victims', 'Number of Latino Female Victims' : 'latino_female_victims', 'Number of Asian Female Victims' : 'asian_female_victims', 'Number of American Indian or Alaska Native Female Victims' : 'amerind_nativealask_female_victims', 'Number of Other Race Female Victims' : 'other_race_female_victims'},
                      inplace = True)
```

```python
# Checking renaming of columns

df_2.columns
```

```python
# Rename columns to be all lowercase for consistency and adding underscore between names

df_2.rename(columns = {'ID' : 'id', 'Execution Volunteer' : 'execution_volunteer', 
                     'Number of Victims' : 'number_victims', 'Juvenile' : 'juvenile',
                     'Race' : 'race', 'Sex' : 'sex', 'Region' : 'region', 'Country' : 'country', 'State' : 'state', 'Foreign National' : 'foreign_national', 'Execution Date' : 'execution_date', 'Victim(s) Race(s)' : 'victims_races'},
                      inplace = True)
```

```python
# Checking renaming of columns

df_2.columns
```

```python
# Checking for mixed-type data

for col in df_2.columns.tolist():
      weird = (df_2[[col]].applymap(type) != df_2[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df_2[weird]) > 0:
        print (col)
```
