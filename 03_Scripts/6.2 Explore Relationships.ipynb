### Creating a correlation heatmap in seaborn

```python
# Create a new subset excluding irrelevant columns

sub = df[['number_victims', 'white_male_victims', 'black_male_victims', 'latino_male_victims','asian_male_victims', 'native_amer_male_victims', 'other_race_male_victims', 'white_female_victims', 'black_female_victims', 'latino_female_victims', 'asian_female_victims', 'amerind_nativealask_female_victims', 'other_race_female_victims']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(13,13))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

#### This heatmap in Seaborn is more useful to me as it applies the numeric value to each variable which is easier to read than just the color scale. This visualization reinforces the relationships from matplotlib. 

    The following groups of victims have the strongest relationship to number_victims:
        white male victims
        white female victims
        other race male victims
        other race female victims
        latino female vicitms 
    This suggests that that there is strong relationship between how many victims a person victimized and their gender and race, white male being the strongest.
